```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#ruta de carpetas
ruta_files_GEO = "/content/drive/MyDrive/Radar_Challenge/Data-Base/Data-Complet/GeoTIFF"
files = []

#rutas de imágenes
rutas_F = []
rutas_G = []
rutas_H = []
rutas_I = []
rutas_J = []
rutas_K = []
rutas_L = []
rutas_M = []
rutas_N = []
rutas_O = []

#extraemos y dividimos las rutas de las imágenes 
for files_p in os.listdir(ruta_files_GEO):
  files.append(files_p)
files.sort()
for lt in files:
  ruta = ruta_files_GEO + '/' + lt
  if lt == 'F':
    for name in os.listdir(ruta):
      rutas_F.append(name)
  if lt == 'G':
    for name in os.listdir(ruta):
      rutas_G.append(name)
  if lt == 'H':
    for name in os.listdir(ruta):
      rutas_H.append(name)
  if lt == 'I':
    for name in os.listdir(ruta):
      rutas_I.append(name)
  if lt == 'J':
    for name in os.listdir(ruta):
      rutas_J.append(name)
  if lt == 'K':
    for name in os.listdir(ruta):
      rutas_K.append(name)
  if lt == 'L':
    for name in os.listdir(ruta):
      rutas_L.append(name)
  if lt == 'M':
    for name in os.listdir(ruta):
      rutas_M.append(name)
  if lt == 'N':
    for name in os.listdir(ruta):
      rutas_N.append(name)
  if lt == 'O':
    for name in os.listdir(ruta):
      rutas_O.append(name)

print(len(rutas_F)+len(rutas_G)+len(rutas_H)+len(rutas_I)+len(rutas_J)+len(rutas_K)+len(rutas_L)+len(rutas_M)+len(rutas_N)+len(rutas_O))
```

```python
import matplotlib.pyplot as plt
from skimage import io
import random
import matplotlib.pyplot as plt
import cv2
import pandas as pd
PATH = '/content/drive/MyDrive/Radar_Challenge/Data-Base/Data_IAC/500'
data = pd.read_csv(PATH + '/Data_Complete_500.csv')
names = data.columns.values
```

```python
path = '/content/drive/MyDrive/Radar_Challenge/Data-Base/Data-Complet/GeoTIFF'
r, c = 2, 5
cont = 0
fig = plt.figure(figsize=(3*c, 3*r))
for _r in range(r):
    for _c in range(c):
        plt.subplot(r, c, _r*c + _c + 1)
        clase = names[cont]
        ix = random.randint(0, data.shape[0]-1)        
        name = data[clase][ix]
        clase = names[cont].split('Class_')[1]
        img = cv2.imread(f'{path}/{clase}/{name}')
        plt.imshow(img)
        label = clase
        plt.title(label)
        plt.axis(False)
        cont+=1
plt.show()
```
